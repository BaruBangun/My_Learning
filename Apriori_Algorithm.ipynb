```python
transactions = []
for i in range(0, 14963):
  transactions.append([str(df.values[i,j]) for j in range(0, 11)])
```

```python
transactions
```

```python
from apyori import apriori
```

```python
results= apriori(transactions= transactions, min_support = 0.0007, min_confidence = 0.1, min_lift = 2, min_length = 2 )
```

```python
hasil=list(results)
```

```python
hasil
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(hasil), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame
```

```python
resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```
