```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
X,y = load_breast_cancer(return_X_y=True, as_frame=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier()
model2.fit(X_train, y_train)
```

```python
from sklearn.svm import SVC
model3 = SVC()
model3.fit(X_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
model4 = GaussianNB()
model4.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
model5 = DecisionTreeClassifier()
model5.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model6 = RandomForestClassifier()
model6.fit(X_train,y_train)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report
```

```python
y_pred1 = model1.predict(SC.transform(X_test))
y_pred2 = model2.predict(SC.transform(X_test))
y_pred3 = model3.predict(SC.transform(X_test))
y_pred4 = model4.predict(SC.transform(X_test))
y_pred5 = model5.predict(SC.transform(X_test))
y_pred6 = model6.predict(SC.transform(X_test))
```

```python
print(classification_report(y_test, y_pred1))
CM = confusion_matrix(y_test, y_pred1)
ConfusionMatrixDisplay(CM).plot()
plt.title('Confusion Matrix Logistic Regression')
plt.show()
```

```python
print(classification_report(y_test, y_pred2))
CM = confusion_matrix(y_test, y_pred2)
ConfusionMatrixDisplay(CM).plot()
plt.title('Confusion Matrix k-NN')
plt.show()
```

```python
print(classification_report(y_test, y_pred3))
CM = confusion_matrix(y_test, y_pred3)
ConfusionMatrixDisplay(CM).plot()
plt.title('Confusion Matrix SVM')
plt.show()
```

```python
print(classification_report(y_test, y_pred4))
CM = confusion_matrix(y_test, y_pred4)
ConfusionMatrixDisplay(CM).plot()
plt.title('Confusion Matrix Naive Bayes')
plt.show()
```

```python
print(classification_report(y_test, y_pred5))
CM = confusion_matrix(y_test, y_pred5)
ConfusionMatrixDisplay(CM).plot()
plt.title('Confusion Matrix Decision Tree')
plt.show()
```

```python
print(classification_report(y_test, y_pred6))
CM = confusion_matrix(y_test, y_pred6)
ConfusionMatrixDisplay(CM).plot()
plt.title('Confusion Matrix Random Forest')
plt.show()
```
