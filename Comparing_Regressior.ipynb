```python
df = df.drop(['Date/Time'], axis=1)
```

```python
df.head(len(df))
```

```python
y = df.iloc[:, 0:1].values
X = df.iloc[:, 1:-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state =1 )
```

```python
y_resh = y_train.reshape(len(y_train),-1)
y_resh_test = y_test.reshape(len(y_test), -1)
```

```python
from sklearn.preprocessing import StandardScaler
SC_X = StandardScaler()
SC_y = StandardScaler()
SC_X_test = StandardScaler()
SC_y_test = StandardScaler()
X_trans = SC_X.fit_transform(X_train)
y_trans = SC_y.fit_transform(y_resh)
X_trans_test = SC_X_test.fit_transform(X_test)
y_trans_test = SC_y_test.fit_transform(y_resh_test)
print(X_trans)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures()
X_poly = pf.fit_transform(X_train)
X_poly_test = pf.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
modelLR = LinearRegression()
modelLR.fit(X_train,y_train)
y_predLR = modelLR.predict(X_test)
```

```python
modelPR = LinearRegression()
modelPR.fit(X_poly,y_train)
y_predPR = modelPR.predict(X_poly_test)
```

```python
from sklearn.svm import SVR
modelSVR = SVR()
modelSVR.fit(X_trans, y_trans)
y_predSVR = modelSVR.predict(X_trans_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
modelDTR = DecisionTreeRegressor()
modelDTR.fit(X_train, y_train)
y_predDTR = modelDTR.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
modelRFR = RandomForestRegressor()
modelRFR.fit(X_train, y_train)
y_predRFR = modelRFR.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_predLR, y_test)
```

```python
r2_score(y_predPR, y_test)
```

```python
r2_score(y_predSVR, y_trans_test)
```

```python
r2_score(y_predDTR, y_test)
```

```python
r2_score(y_predRFR, y_test)
```
